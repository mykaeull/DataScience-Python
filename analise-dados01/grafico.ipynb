```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(0, 10, 0.1)
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 10))
fig1.suptitle('Primeiro gráfico', size=30)

f1_axes[0, 0].plot(np.sin(x))
f1_axes[0, 0].set_title('Caixa 1')
f1_axes[0, 0].set_xlabel('eixo x', size=15)
f1_axes[0, 0].set_ylabel('eixeo y', size=15)

f1_axes[0, 1].plot(np.cos(x))
f1_axes[0, 1].set_title('Caixa 2')
f1_axes[0, 1].set_xlabel('eixo x', size=15)
f1_axes[0, 1].set_ylabel('eixo y', size=15)

f1_axes[1, 0].plot(np.tan(x))
f1_axes[1, 0].set_title('Caixa 3')
f1_axes[1, 0].set_xlabel('eixo x', size=15)
f1_axes[1, 0].set_ylabel('eixo y', size=15)

f1_axes[1, 1].plot(np.sin(x), label='sen(x)')
f1_axes[1, 1].plot(np.cos(x), label='cos(x)')
f1_axes[1, 1].set_title('Caixa 4')
f1_axes[1, 1].set_xlabel('eixo x', size=15)
f1_axes[1, 1].set_ylabel('eixo y', size=15)
f1_axes[1, 1].legend()
```
