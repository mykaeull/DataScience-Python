```python
import pandas as pd
```

```python
colunas = ['NU_INSCRICAO', 'NU_ANO', 'CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA', 'CO_UF_RESIDENCIA',
           'SG_UF_RESIDENCIA', 'NU_IDADE', 'TP_SEXO', 'TP_COR_RACA', 'IN_GESTANTE', 'NU_NOTA_REDACAO', 'NU_NOTA_MT', 'NU_NOTA_CH', 
          'NU_NOTA_CN', 'NU_NOTA_LC']
microdadosEnem = pd.read_csv("C:/Users/mykaeull/Desktop/DataScience/analise-dados/DADOS_ENEM/DADOS/MICRODADOS_ENEM_2018.csv",
                             sep=";", encoding="ISO-8859-1", usecols=colunas, nrows=100000)
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items=['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA',
                                                         'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA', 'NU_NOTA_REDACAO',
                                                         'NU_NOTA_MT', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC'])
microdadosEnemSelecionado.head()
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_RESIDENCIA',
                                                         'CO_UF_RESIDENCIA', 'SG_UF_RESIDENCIA'], as_index=False).mean()
```
