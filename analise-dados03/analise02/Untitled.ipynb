```python
import pandas as pd
```

```python
dados = pd.read_csv('sinasc122017.csv', sep=';')
```

```python
dados
```

```python
import matplotlib.pyplot as plt
```

```python
dadosMesPartoVaginal = dados.filter(items=['ano_mes', 'vaginal'])
dadosMesPartoVaginal
```

```python
plt.bar(dadosMesPartoVaginal['ano_mes'], dadosMesPartoVaginal['vaginal'])
plt.xlabel('Mês')
plt.ylabel('Parto Vaginal')
plt.show()
```

```python
plt.barh(dados['ano_mes'], dados['cesario'])
plt.xlabel('Parto Cesariano')
plt.ylabel('Mês')
plt.show()
```

```python
totalPartoVaginal = dados['vaginal'].sum()
totalPartoCesario = dados['cesario'].sum()
totalDePartos = totalPartoVaginal + totalPartoCesario
```

```python
percentualDePartoVaginal = int((100*totalPartoVaginal)/totalDePartos)
percentualDePartoCesario = int((100*totalPartoCesario)/totalDePartos)
```

```python
print(f'Porcentagem de partos vaginais no ano de 2017: {percentualDePartoVaginal}% (valor aproximado)')
print(f'Porcentagem de partos casarianos no ano de 2017: {percentualDePartoCesario}% (valor aproximado)')
```
