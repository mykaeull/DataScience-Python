# Dados sobre o covid-19 no CE

ANÁLISE DOS CONTÁGIOS EM FORTALEZA

```python
dadosCE = dados.loc[dados.state == 'CE']
```

```python
dadosCE
```

```python
dadosCEFortaleza = dadosCE.loc[dadosCE.city == 'Fortaleza']
```

```python
dadosCEFortaleza
```

```python
contagioPeloCovidFortalezaCE = dadosCEFortaleza['confirmed'].sum()
print(f'Número de contágios pelo covid-19 em Fortaleza CE: {contagioPeloCovidFortalezaCE}')
```

```python
contagioPeloCovidCE = dadosCE['confirmed'].sum()
print(f'Número de contágios pelo covid-19 no Estado do Ceará: {contagioPeloCovidCE}')
```

```python
percentualDeContagioFortalezaCompCE = int((100*contagioPeloCovidFortalezaCE)/contagioPeloCovidCE)
```

```python
print(f'Percentual de contágios em Fortaleza comparado ao Estado(CE): {percentualDeMortesFortalezaCompCE}% (valor aproximado)')
```

```python
qtdCidadesCE = dadosCE['city'].count()
print(f'Quantidade de cidades do CE: {qtdCidadesCE}')
```

ANÁLISE DAS MORTES EM FORTALEZA

```python
qtdMortesCE = dadosCE['deaths'].sum()
print(f'Quantidade de mortes causadas pelo covid-19 no Estado do Ceará: {qtdMortesCE}')
```

```python
qtdMortesFortalezaCE = dadosCEFortaleza.deaths.sum()
print(f'Número de mortes causadas pelo covid-19 em Fortaleza: {qtdMortesFortalezaCE}')
```

```python
percentualMortesFortalezaCompCE = int((100*qtdMortesFortalezaCE)/qtdMortesCE)
print(f'Percentual de mortes pelo covid-19 em Fortaleza em comparação ao Estado do Ceará: {percentualMortesFortalezaCompCE}% (valor aproximado)')
```

# Gráfico(s) das relaçoes de contágio e mortes no Brasil

```python
dados
```

```python
dadosEstadoContagio = dados.filter(items=['state', 'confirmed']).groupby('state', as_index=False).sum()
dadosEstadoContagio
```

GRÁFICO DE CONTÁGIO POR ESTADO 

```python
dadosEstadoContagio = dados.filter(items=['state', 'confirmed']).groupby('state').sum().plot.bar()
```

GRÁFICO DE MORTES POR ESTADO

```python
dadosEstadoMortes = dados.filter(items=['state', 'deaths']).groupby('state').sum().sort_values(by='deaths', ascending=False)
dadosEstadoMortes
```

```python
dados.filter(items=['state', 'deaths']).groupby('state').sum().plot.bar()
```
